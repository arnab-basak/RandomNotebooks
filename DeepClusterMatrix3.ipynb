```python
import torch
from torch import nn
from torch.optim import Adam
import torch.optim as optim
from torch.utils.data import DataLoader
```

```python
data_loader=torch.utils.data.DataLoader(dataset=arr_train,batch_size=32,shuffle=True)
```

```python
class AutoEncoder(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = torch.nn.Sequential(
            torch.nn.Linear(64, 32),
            torch.nn.LeakyReLU(),
            torch.nn.Linear(32, 16),
            torch.nn.LeakyReLU(),
            torch.nn.Linear(16, 8),
            torch.nn.ReLU(),
            torch.nn.Linear(8, 4),
            torch.nn.ReLU(),
            torch.nn.Linear(4, 2),
            #torch.nn.ReLU(),
            #torch.nn.Linear(6, 5),
            #torch.nn.ReLU(),
            #torch.nn.Linear(5, 4),
            #torch.nn.ReLU(),
            #torch.nn.Linear(4, 3),
        )

        self.decoder = torch.nn.Sequential(
            torch.nn.Linear(2, 4),
            torch.nn.ReLU(),
            torch.nn.Linear(4, 8),
            torch.nn.ReLU(),
            torch.nn.Linear(8, 16),
            torch.nn.LeakyReLU(),
            torch.nn.Linear(16, 32),
            torch.nn.LeakyReLU(),
            torch.nn.Linear(32, 64),
            #torch.nn.ReLU(),
            #torch.nn.Linear(8, 9),
            #torch.nn.ReLU(),
            #torch.nn.Linear(9, 10),
            #torch.nn.ReLU(),
            #torch.nn.Linear(10, 11),
            #torch.nn.Sigmoid()
        )
    def encode(self, x):
        encoded = self.encoder(x)
        return encoded

    def decode(self, x):
        decoded = self.decoder(x)
        return decoded
 
    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded
```

```python
learning_rate=0.001
model=AutoEncoder()
mse=nn.MSELoss()
optimizer=torch.optim.Adam(model.parameters(),lr=learning_rate)
scheduler=optim.lr_scheduler.ReduceLROnPlateau(optimizer,patience=5,verbose=True)
```

```python
import math
num_epochs=150
total_samples= len(arr_train)
n_iterations=math.ceil(total_samples/32)
print(f'total {total_samples} n_iterations {n_iterations}')
```

```python
outputs=[]
L=[]
for epoch in range(num_epochs):
    for step, x in enumerate(data_loader):
        recon=model(x.float())
        loss=mse(recon,x.float())
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    L.append(loss.item())
    mean_loss=sum(L)/len(L)
    scheduler.step(mean_loss)
    print(f'Epoch:{epoch+1}, Loss:{loss.item():.6f}')
    outputs.append((epoch,x,recon))
```

```python
import matplotlib.pyplot as plt
plt.plot(L)
```

```python
data_loader=DataLoader(dataset=arr_train,batch_size=len(arr_train))
```

```python
for step, x in enumerate(data_loader):
    recon1=model.encode(x.float())
```

```python
data_loader=DataLoader(dataset=arr_test,batch_size=len(arr_test))
```

```python
for step, x in enumerate(data_loader):
    recon2=model.encode(x.float())
```

```python
class AutoEncoder(Model):
    def __init__(self):
        super(AutoEncoder,self).__init__()
        
        self.encoder = tf.keras.Sequential([
            #tf.keras.Input(shape=(None,64)),
            tf.keras.layers.InputLayer(input_shape=(64,)),
            Dense(24,activation=LeakyReLU(alpha=0.01),),
            BatchNormalization(),
            #Dropout(0.2),
            Dense(12,activation=LeakyReLU(alpha=0.01)),
            #Dropout(0.2),
            BatchNormalization(),
            Dense(6,activation=LeakyReLU(alpha=0.01)),
            #Dropout(0.2),
            BatchNormalization(),
            #Dense(4,activation=LeakyReLU(alpha=0.01)),
            #Dropout(0.1),
            Dense(2),
            BatchNormalization(),]
        )

        self.decoder = tf.keras.Sequential([
            Dense(2,activation=LeakyReLU(alpha=0.01)),
            #Dropout(0.2),
            BatchNormalization(),
            Dense(6,activation=LeakyReLU(alpha=0.01)),
            #Dropout(0.2),
            BatchNormalization(),
            Dense(12,activation=LeakyReLU(alpha=0.01)),
            #Dropout(0.2),
            BatchNormalization(),
            Dense(24,activation=LeakyReLU(alpha=0.01)),
            #Dropout(0.2),
            BatchNormalization(),
            Dense(64),]
        )
    def encode(self, x):
        encoded = self.encoder(x)
        return encoded

    def decode(self, x):
        decoded = self.decoder(x)
        return decoded
 
    def call(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded
```

```python
class MyCallback(tf.keras.callbacks.Callback):

    def __init__(self, patience=0):
        super(MyCallback, self).__init__()

    def on_epoch_begin(self, epoch, logs=None):
        tf.print(self.model.get_weights())
```
